# Executive Summary
This notebook presents a professional Exploratory Data Analysis (EDA) of the Iris Dataset using Python.


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Load Dataset

```python
iris = pd.read_csv("Iris.csv")
print("Dataset Loaded Successfully!")
```

# Dataset Inspection

```python
print("SHAPE:", iris.shape)
print("\nCOLUMNS:")
print(iris.columns)
print("\nFIRST FIVE ROWS:")
iris.head()
```

## Observation
- Dataset contains 150 rows and 6 columns.
- Four numerical features and one categorical feature.


# Dataset Information

```python
iris.info()
```

```python
iris.describe()
```

# Missing Values Check

```python
iris.isnull().sum()
```

# Scatter Plot Analysis

```python
plt.figure(figsize=(10,5), )

sns.scatterplot(data=iris, x='SepalLengthCm', y='PetalLengthCm', hue='Species', s=80, palette=["#065186", "#bf5f0b", "#0c9d0c"])

plt.title('Sepal Length vs Petal Length', fontsize=20, fontweight='bold', color='#1f3b73')
plt.xlabel('Sepal Length (cm)', fontsize=12, color='#2c3e50')
plt.ylabel('Petal Length (cm)', fontsize=12, color='#2c3e50')


plt.show()
```

```python
plt.figure(figsize=(10,5), )

sns.scatterplot(data=iris, x='SepalWidthCm', y='PetalWidthCm', hue='Species', s=80, palette=["#065186", "#bf5f0b", "#0c9d0c"])

plt.title('Sepal Width vs Petal Width', fontsize=20, fontweight='bold', color='#1f3b73')
plt.xlabel('Sepal Width (cm)', fontsize=12, color='#2c3e50')
plt.ylabel('Petal Width (cm)', fontsize=12, color='#2c3e50')


plt.show()
```

### Insight
- Setosa forms a separate cluster.
- Positive relationship exists between sepal length and petal length  or , sepal width and petal width,


# Histogram Analysis

```python
plt.figure(figsize=(10,5))

sns.histplot(iris['SepalLengthCm'],bins=10,kde=True, color="#4141A5")

plt.title('Distribution of Sepal Length',fontsize=20,fontweight='bold',color='#1f3b73')

plt.xlabel('Sepal Length (cm)', color='#2c3e50')
plt.ylabel('Frequency', color='#2c3e50')

plt.show()
```

### Insight
Most flowers have sepal lengths between 5 and 7 cm.


# Box Plot Analysis

```python
plt.figure(figsize=(10,5))

sns.boxplot(data=iris,x='Species',y='PetalLengthCm',hue="Species",
            palette=["#065186", "#bf5f0b", "#0c9d0c"])

plt.title( 'Petal Length Distribution by Species',fontsize=20,fontweight='bold', color='#1f3b73')

plt.xlabel('Species', fontsize=12, color='#2c3e50')
plt.ylabel('Petal Length (cm)', fontsize=12, color='#2c3e50')

plt.show()
```

### Insight
Virginica has the highest petal lengths while Setosa has the smallest.


# Correlation Heatmap

```python
plt.figure(figsize=(10,5))
sns.heatmap(
    iris.drop('Species', axis=1).corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title('Correlation Heatmap')
plt.show()
```

# Key Findings

1. No missing values were found.
2. Petal Length and Petal Width are strongly correlated.
3. Setosa is clearly distinguishable from other species.
4. Petal features are highly useful for classification.
